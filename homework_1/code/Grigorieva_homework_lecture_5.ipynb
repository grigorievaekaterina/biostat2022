Напишите функцию, которая будет принимать на вход экспрессии гена, семплировать их них выборки размера `sample_size`, считать среднюю экспрессию и повторять это `n_samples` раз. Примените эту функцию к экспрессиям гена `TMCC1` в обоих клеточных типах, визуализируйте их. Отличаются ли средние экспрессии данного гена у этих клеточных типов?
 
Сигнатура функции:

```python
def demonstrate_clt(expressions, sample_size, n_samples):
    mean_expressions = []

    return mean_expressions
```

````python
def demonstrate_clt(expressions, sample_size, n_samples):
    mean_expressions = []
    for i in range(n_samples):
      sample = expressions.sample(sample_size)
      mean_expressions.append(np.mean(sample))

    return mean_expressions
````

````python
sample_size = 200
n_samples = 1000
````

````python
#b-cells
mean_b_cells = demonstrate_clt(b_cells_expression_data[example_gene], sample_size, n_samples)
sns.histplot(mean_b_cells, stat="density")
````

````python
#nk-cells
mean_nk_cells = demonstrate_clt(nk_cells_expression_data[example_gene], sample_size, n_samples)
sns.histplot(mean_nk_cells, stat="density")
````

А теперь посчитайте 95% доверительные интервалы для обоих распределений (примем тот факт, что средние распределены нормально для обоих клеточных типов) и скажите, отличается ли средняя экспрессия данного гена между клеточными типами?

````python
def calculate_ci(expressions, sample_size, n_samples):
    std = np.std(expressions)
    mean = np.mean(demonstrate_clt(expressions, sample_size, n_samples))
    se = std / np.sqrt(sample_size)
    left_b = mean - 1.96 * se
    right_b = mean + 1.96 * se

    return left_b, right_b
````

````python
#b-cells
print(np.mean(b_cells_expression_data[example_gene]))
print(calculate_ci(b_cells_expression_data[example_gene], sample_size, n_samples))
````

````python
#nk-cells
print(np.mean(nk_cells_expression_data[example_gene]))
print(calculate_ci(nk_cells_expression_data[example_gene], sample_size, n_samples))
````

*Средняя экспрессия по всей данной выборке НК клеток выше, чем у B клеток, однако 95% доверительные интервалы пересекаются, поэтому в целом мы не можем так утверждать.*

````python
def check_intervals_intersect(first_ci, second_ci):   
    if first_ci[0] > second_ci[1] or first_ci[1] < second_ci[0]:
      are_intersect = False
    else:
      are_intersect = True
    return are_intersect # True or False
````

````python
def check_dge_with_ci(first_table, second_table):
    # dge - differential gene expression
    columns = list(first_table)
    ci_test_results = []
    for i in columns:
        ci_first = st.t.interval(alpha=0.95, df=len(first_table[i]) - 1, 
                             loc=np.mean(first_table[i]), 
                             scale=st.sem(first_table[i]))
        ci_second = st.t.interval(alpha=0.95, df=len(second_table[i]) - 1, 
                             loc=np.mean(second_table[i]), 
                             scale=st.sem(second_table[i]))
        ci_test_results.append(check_intervals_intersect(ci_first, ci_second))

    return ci_test_results
````

````python
def check_dge_with_ztest(first_table, second_table):
    # dge - differential gene expression
    columns = list(first_table)
    z_test_results = []
    for i in columns:
        ztest = ztest(first_table[i], second_table[i])
        if ztest[1] < 0.05:
            z_test_results.append(True)
        else:
            z_test_results.append(False)

    return z_test_results
````
